```python
import numpy as np

print("=== Topic 1: From Lists (np.array) ===")
arr1 = np.array([1, 2, 3, 4])
print(arr1)

print("\n=== Topic 2: From Tuples ===")
arr2 = np.array((10, 20, 30))
print(arr2)

print("\n=== Topic 3: From Range (np.arange) ===")
arr3 = np.arange(1, 10, 2)   # start, stop, step
print(arr3)

print("\n=== Topic 4: Zeros Array (np.zeros) ===")
arr4 = np.zeros((2, 3))
print(arr4)

print("\n=== Topic 5: Ones Array (np.ones) ===")
arr5 = np.ones((2, 2))
print(arr5)

print("\n=== Topic 6: Empty Array (np.empty) ===")
arr6 = np.empty((3, 3))
print(arr6)

print("\n=== Topic 7: Full Array (np.full) ===")
arr7 = np.full((3, 3), 7)   # full of 7
print(arr7)

print("\n=== Topic 8: Identity Matrix (np.eye) ===")
arr8 = np.eye(4)
print(arr8)

print("\n=== Topic 9: Diagonal Matrix (np.diag) ===")
arr9 = np.diag([5, 10, 15])
print(arr9)

print("\n=== Topic 10: Linspace (np.linspace) ===")
arr10 = np.linspace(0, 1, 5)
print(arr10)

print("\n=== Topic 11: Logspace (np.logspace) ===")
arr11 = np.logspace(1, 3, 5)  # 10^1 to 10^3
print(arr11)

print("\n=== Topic 12: From Existing Data (np.asarray) ===")
my_list = [100, 200, 300]
arr12 = np.asarray(my_list)
print(arr12)

print("\n=== Topic 13: From Function (np.fromfunction) ===")
# matrix[i,j] = i + j
arr13 = np.fromfunction(lambda i, j: i + j, (3, 3), dtype=int)
print(arr13)

print("\n=== Topic 14: Random Arrays (np.random.rand) ===")
arr14 = np.random.rand(3, 3)
print(arr14)

print("\n=== Topic 15: Structured Arrays Intro ===")
data_type = [('name', 'U10'), ('age', 'i4'), ('marks', 'f4')]
students = np.array([("Ali", 21, 88.5), ("Sara", 19, 92.0)], dtype=data_type)
print(students)

print("\n=== Topic 16: Array with Custom Dtype ===")
custom = np.array([5, 10, 15], dtype='float32')
print(custom)
```
